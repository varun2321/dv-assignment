```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load education data from a CSV file
education_data = pd.read_csv('/content/sample_data/Student_Marks.csv')




# 2. Visualize the distribution of grades using violin plots
plt.figure(figsize=(10, 6))
sns.violinplot(x='time_study', y='Marks', data=education_data)
plt.xlabel('time_study')
plt.ylabel('Marks')
plt.title('Distribution (Violin Plot)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()





```
